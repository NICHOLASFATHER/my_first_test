```python
import pandas as pd 
import numpy as np 
from tqdm import tqdm
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',100)
pd.options.mode.chained_assignment = None
```

```python
"""
clockAtArrival: 16-digit timestamp
StockID: StockID 
time: time 
ID_map: for plotting, matching the stockID 
snapmark: for the same snapmark, it means the stocks info belongs in one snapshot 
ms_caa: more reable timestamp
"""
snaps = pd.read_csv("md2020.csv", index_col=0)
snaps.head(100)
```

```python
stock_id = snaps['StockID'].unique()
stock_id.sort()
```

```python
stock_id
```

```python
list_A = [[], [], [], []]
```

```python
def is_sorted_single_snapmark(snapmark_id, n_list):
    df_snapmark = snaps[snaps['snapmark'] == snapmark_id]
    temp_df = pd.DataFrame(n_list, columns=['StockID'])
    temp_df = temp_df.merge(df_snapmark, on='StockID')
    temp_df.sort_values(by='StockID')
#     print(temp_df)
    for i in range(len(temp_df['clockAtArrival'])-1):
        if temp_df['clockAtArrival'].iloc[i]>temp_df['clockAtArrival'].iloc[i+1]:
            return False
    return True
```

```python
unique_snap = snaps['snapmark'].unique()
for i in tqdm(stock_id):
    for l_num in range(4):
        list_A[l_num].append(i)
        for snap_id in unique_snap:
            if not is_sorted_single_snapmark(snap_id, list_A[l_num]):
                list_A[l_num].pop()
                break
        if snap_id == 5211:
            break
```

```python
group_1 = pd.DataFrame({'StockID':list_A[0], 'Group_No':1})
group_2 = pd.DataFrame({'StockID':list_A[1], 'Group_No':2})
group_3 = pd.DataFrame({'StockID':list_A[2], 'Group_No':3})
group_4 = pd.DataFrame({'StockID':list_A[3], 'Group_No':4})
group_ans = pd.concat([group_1, group_2, group_3, group_4], axis=0)
```

```python
group_ans.to_csv('2020_ans.csv')
```

```python
def is_decrease(test_ID):
    ## create the dataframe with specific snapmark_ID
    test = snaps[snaps['snapmark'] == test_ID]
    test = test.merge(group_ans, on='StockID')
    ## pre-process the data
    test.sort_values(by = ['Group_No', 'clockAtArrival'],ascending = True, inplace = True)  # 按时间递增排序
    grouped = test.groupby('Group_No')
    for i in range(len(grouped)):
        try:
            grouped_data = grouped.get_group(i+1)
        except KeyError:
            continue
        grouped_data.reset_index(drop=True, inplace=True)
        for j in range(len(grouped_data['clockAtArrival'])-1):
            if grouped_data['StockID'].iloc[j] >  grouped_data['StockID'].iloc[j+1]:
                return False
    return True
```

```python
i = 0
for snapmark_id in tqdm(snaps['snapmark'].unique()):
    temp_result = is_decrease(snapmark_id)
    i += 1
    if not temp_result:
        print(f'{snapmark_id} miss_match')
        break
    if i == len(snaps['snapmark'].unique()-1):
        print('match well!')
```
