```python
"""
For MD2019, the groupping pattern is same for all snapmarks. 
Here's the dataframe illustrate each StockID belongs to which group. 

All 2128 StockID are divided into 4 groups. 
"""
grouping_ans = pd.read_csv("answers.csv")
grouping_ans
```

```python
test100 = snaps[snaps['snapmark']==100]
test100
```

```python
test100 = test100.merge(grouping_ans, on='StockID')
```

```python
test100.sort_values(['Group_No', 'clockAtArrival'], inplace=True)
```

```python
test100
```

```python
grouped = test100.groupby('Group_No')
```

```python
len(grouped)
```

```python
grouped_data = grouped.get_group(1)
```

```python
def is_decrease(test_ID):
    ## create the dataframe with specific snapmark_ID
    test = snaps[snaps['snapmark'] == test_ID]
    test = test.merge(grouping_ans, on='StockID')
    ## pre-process the data
    test.sort_values(by = ['Group_No', 'clockAtArrival'],ascending = True, inplace = True)  # 按时间递增排序
    grouped = test.groupby('Group_No')
    for i in range(len(grouped)):
        try:
            grouped_data = grouped.get_group(i+1)
        except KeyError:
            continue
        grouped_data.reset_index(drop=True, inplace=True)
        for j in range(len(grouped_data['clockAtArrival'])-1):
            if grouped_data['StockID'].iloc[j] >  grouped_data['StockID'].iloc[j+1]:
                return False
    return True
```

```python
is_decrease(100)
```

```python
i = 0
for snapmark_id in tqdm(snaps['snapmark'].unique()):
    temp_result = is_decrease(snapmark_id)
    i += 1
    if not temp_result:
        print(f'{snapmark_id} miss_match')
        break
    if i == len(snaps['snapmark'].unique()-1):
        print('match well!')
```
